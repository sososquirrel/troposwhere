# Simulation Check

Quick sanity check: load split_1, inspect the raw datasets (1D, 2D, 3D) and computed variables.

```python
import sys
sys.path.insert(0, '/Users/sophieabramian/Documents/troposwhere')

from tropokit.Simulation import Simulation

PATH_RAW_DATA = '/Users/sophieabramian/Documents/troposwhere/data/squall_lines'
PATH_SAVED    = '/Users/sophieabramian/Documents/troposwhere/data/squall_lines/saved_simu'

SPLIT = 1  # change this to check other splits

paths = {
    'path_1d': f'{PATH_RAW_DATA}/1D/split_{SPLIT}.nc',
    'path_2d': f'{PATH_RAW_DATA}/2D/split_{SPLIT}.nc',
    'path_3d': f'{PATH_RAW_DATA}/3D/split_{SPLIT}.nc',
}

simu = Simulation(
    data_folder_paths=[paths['path_1d'], paths['path_2d'], paths['path_3d']],
    velocity='8',
    temperature='300',
    bowen_ratio='1',
    microphysic='1',
    split=str(SPLIT),
)

print(f"Loaded: {simu.name}")
print(f"  Time steps : {simu.nt}")
print(f"  Grid       : {simu.nz}z × {simu.ny}y × {simu.nx}x")
```

## Raw datasets

```python
print("=== dataset_1d ===")
display(simu.dataset_1d)

print("=== dataset_2d ===")
display(simu.dataset_2d)

print("=== dataset_3d ===")
display(simu.dataset_3d)
```

## Load saved simulation and display computed datasets

```python
simu.load(backup_folder_path=PATH_SAVED)

print("=== dataset_computed_3d ===")
display(simu.dataset_computed_3d)

print("=== dataset_computed_2d ===")
display(simu.dataset_computed_2d)

print("=== dataset_isentropic ===")
display(simu.dataset_isentropic)
```
